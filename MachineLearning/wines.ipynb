```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
data = load_wine()['data']
target = load_wine()['target']
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=load_wine()['feature_names'])
df['target'] = target
```

```python
df.info()
```

```python
df.describe() # 도메인 지식이 없으니 기초 통계 결과를 봐도 인사이트를 얻을 수 없다.
```

```python
df.groupby('target').mean()
```

```python
# train_data_split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, stratify=target)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': svm.SVC(),
    'SGD Classifier': SGDClassifier(),
    'Logistic Regression': LogisticRegression()
}

for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f'{model_name} Report:')
    print(classification_report(y_test, y_pred))


# RandomForest가 가장 좋은 성능을 보인다.
# precision과 recall의 조화평균인 f-1 score가 높다.

```
