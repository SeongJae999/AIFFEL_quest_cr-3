```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!


menu = ['HOT americano', 'ICE latte', 'ICE mocha', 'ICE choco_latte']
price = [2000, 3000, 3000, 3000]

class Kiosk:
    def __init__(self, menu, price):
        self.menu = menu
        self.price = price
        self.order_list = []
        self.price_sum = 0

    # 메뉴 출력 메서드
    def menu_print(self):
        for x, (y, z) in enumerate(zip(self.menu, self.price)):
            print(f"{x+1}번 - {y}: {z}원")

    # 주문 메서드
    def menu_select(self):

        while True:

            try:
                order = int(input("원하시는 메뉴 번호를 입력해주세요.: "))

                self.order_list.append((self.menu[order-1], self.price[order-1]))
                self.price_sum += price[order-1]

                print(f"{self.menu[order-1]}: {self.price[order-1]} \n 합계:{self.price_sum}")

                break

            except:
                print("메뉴판에서 번호를 선택해주세요.")

        add_order = ''
        while add_order != 'n':
            try:
                add_order = input('추가 주문 하시겠습니까? (y, n): ')

                if add_order == 'y':

                    order = int(input("원하시는 메뉴 번호를 입력해주세요.: "))

                    self.order_list.append((self.menu[order-1], self.price[order-1]))
                    self.price_sum += price[order-1]

                    print(f"{self.menu[order-1]}: {self.price[order-1]} \n 합계:{self.price_sum}")

                else:
                    print('y 또는 n 중에 선택해주세요.')

            except:
                print("메뉴판에서 번호를 선택해주세요.")


    # 지불
    def pay(self):
        payment = input("지불 방식을 선택해주세요. (card, cash): ")

        if payment in ['cash', '1', '현금']:
            print("직원을 부르겠습니다. 잠시만 기다려주십시오.")

        elif payment in ['card', '2', '카드']:
            print('IC칩 방향에 맞게 카드를 꽂아주세요')

        else:
            print('올바른 지불방식을 입력해주세요.')


    # 주문서 출력
    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_list)):
            print(' ', self.order_list[i][0], ' : ', self.order_list[i][1])

        print(' ', '='*30)

        print('합계 금액 :', self.price_sum)

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')


a = Kiosk(menu, price)  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
