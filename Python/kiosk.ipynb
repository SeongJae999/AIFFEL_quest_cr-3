```python
import customtkinter as ctk
import tkinter as tk
```

```python
class ScrollableFrame(ctk.CTkScrollableFrame):
    def __init__(self, master, title, values):
        super().__init__(master, label_text=title)
        self.values = values
        self.checkboxes = []

        for i, value in enumerate(self.values):
            checkbox = ctk.CTkCheckBox(self, text=value)
            checkbox.grid(row=i, column=0, padx=10, pady=(10,0), sticky='w')
            self.checkboxes.append(checkbox)
            
    def get(self):
        checked_checkboxes = []
        for checkbox in self.checkboxes:
            if checkbox.get() == 1:
                checked_checkboxes.append(checkbox.cget("text"))
        return checked_checkboxes

class Kiosk(ctk.CTk):
    def __init__(self, todaysMenu):
        super().__init__()

        self.title('Kiosk')
        #self.geometry("400x200")

        self.menu = todaysMenu.keys()
        self.price = todaysMenu.values()

        self.basket = {}
        self.price_sum = 0

        self.TodayMenu = ScrollableFrame(self, title='Menu', values=menu)
        self.TodayMenu.grid(row=0, column=0, padx=10, pady=(10, 0), sticky='nsew')

        self.Requirements = ScrollableFrame(self, title='Requirements', values=['HOT', 'ICE'])
        self.Requirements.grid(row=0, column=1, padx=10, pady=(10,0), sticky='nsew')

        self.Cups = ScrollableFrame(self, title='#', values=range(1, 10))
        self.Cups.grid(row=0, column=2, padx=10, pady=(10,0), sticky='nsew')

        self.add_order = ctk.CTkButton(self, text='Add', command=self.add_order)
        self.add_order.grid(row=3, column=0, padx=10, pady=10, sticky='ew', columnspan=1)

        self.check_basket = ctk.CTkButton(self, text='Basket', command=self.check_basket)
        self.check_basket.grid(row=3, column=1, padx=10, pady=10, sticky='ew', columnspan=1)
    
        self.order = ctk.CTkButton(self, text='Order', command=self.order)
        self.order.grid(row=3, column=2, padx=10, pady=10, sticky='ew', columnspan=1)

    def order(self):
        # 결제 방식 선택
        # 주문서 출력

    def add_order(self):
        try:
            order = ((self.TodayMenu.get()[0], self.Requirements.get()[0]), self.Cups.get()[0])
    
            if order[0] not in self.basket:
                self.basket[order[0]] = order[1]
            else:
                self.basket[order[0]] += order[1]
    
            self.price_sum += todaysMenu[order[0][0]] * order[1]
        except:
            pass

    def check_basket(self):
        # 장바구니에서 품목 제거하기 기능 추가
        
        new_window = ctk.CTk()  
        new_window.title("장바구니")
        #new_window.geometry("300x200")
        
        # 장바구니 프레임
        self.cart_frame = ctk.CTkFrame(new_window)
        self.cart_frame.grid(row=0, column=0)

        self.cart_label = ctk.CTkLabel(self.cart_frame, text="장바구니", font=("Helvetica", 18))
        self.cart_label.grid(row=0, column=0)

        # 장바구니 텍스트 박스 크기 조정
        self.cart_listbox = ctk.CTkTextbox(self.cart_frame, height=80, width=500)
        self.cart_listbox.grid(row=1, column=0)

        for menu, count in self.basket.items():
            line = f"{menu[1]+ ' ' +menu[0].upper()} - {count}잔 - {count * todaysMenu[menu[0]]}원\n"
            self.cart_listbox.insert(ctk.END, line)

        self.total_label = ctk.CTkLabel(self.cart_frame, text=f"총 금액: {self.price_sum}원", font=("Helvetica", 18))
        self.total_label.grid(row=3, column=0)
        
        button = ctk.CTkButton(new_window, text="닫기", command=new_window.destroy)
        button.grid(row=4, column=0)
        
        new_window.mainloop()

        
    
        

        
        
```

```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

todaysMenu = {x:y for x, y in zip(menu, price)}

app=Kiosk(todaysMenu)
app.mainloop()
```
