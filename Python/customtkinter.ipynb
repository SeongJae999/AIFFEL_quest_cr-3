```python
import customtkinter as ctk
```

```python
class App(ctk.CTk):
    def __init__(self):
        super().__init__()

        self.title("my app")
        self.geometry("500x300")

        self.status = 0

        self.button_frame = ctk.CTkFrame(self)
        self.button_frame.grid(row=0, column=0, padx=10, pady=(10,0))

        self.button_rect = ctk.CTkButton(self.button_frame, text='직사각형', command=self.rect)
        self.button_rect.grid(row=1, column=0, padx=10, pady=(10,0), sticky="w")
        
        self.button_par = ctk.CTkButton(self.button_frame, text='평행사변형', command=self.par)
        self.button_par.grid(row=2, column=0, padx=10, pady=(10,0), sticky="w")
        
        self.button_trape = ctk.CTkButton(self.button_frame, text='사다리꼴', command=self.trape)
        self.button_trape.grid(row=3, column=0, padx=10, pady=(10,0), sticky="w")

        self.entry = ctk.CTkEntry(self)
        self.entry.grid(row=4, column=0, padx=10, pady=10, sticky="ew")

        self.button_entry = ctk.CTkButton(self, text="입력", command=self.get_value)
        self.button_entry.grid(row=5, column=0, padx=10, pady=10, sticky='ew')

        self.label_answer = ctk.CTkLabel(self)
        self.label_answer.grid(row=5, column=1, padx=10, pady=10, sticky='ew')

        self.button_quit = ctk.CTkButton(self, text="종료", command=lambda: self.destroy())
        self.button_quit.grid(row=6, column=0, padx=10, pady=10)
    
    def rect(self):
        self.status = 1
        text = "가로, 세로를 입력해주세요. (구분자: ,)"
        self.label_answer.configure(text=text)

    def par(self):
        self.status = 2
        text = "밑변과 높이를 입력해주세요. (구분자: ,)"
        self.label_answer.configure(text=text)
        
    def trape(self):
        self.status = 3
        text = "윗변, 밑변, 그리고 높이를 입력해주세요. (구분자: ,)"
        self.label_answer.configure(text=text)

    def get_value(self):

        if self.status == 0:
            text = "우선, 원하는 사각형의 종류를 선택해주세요!"
            self.label_answer.configure(text=text)
        
        if self.status == 1:
            try:
                width, vertical = map(int, self.entry.get().split(','))
                area = width * vertical
                result = '직사각형의 넓이는: ' + str(area)
                self.label_answer.configure(text=result)
            except ValueError:
                text = "ValueError, 다시 시도해주세요!"
                self.label_answer.configure(text=text)
        elif self.status == 2:
            try:
                base, height = map(int, self.entry.get().split(','))
                area = base * height
                result = '평행사변형의 넓이는: ' + str(area)
                self.label_answer.configure(text=result)
            except ValueError:
                text = "ValueError,  다시 시도해주세요!"
                self.label_answer.configure(text=text)
        elif self.status == 3:
            try:
                upper_base, lower_base, height = map(int, self.entry.get().split(','))
                area = (upper_base+lower_base)*height/2
                result = "사다리꼴의 넓이는: " + str(area)
                self.label_answer.configure(text=result)
            except ValueError:
                text = "ValueError,  다시 시도해주세요!"
                self.label_answer.configure(text=text)
        else:
            pass
```

```python
app = App()
app.mainloop()
```

```python
import customtkinter as ctk

class KioskApp(ctk.CTk):
    def __init__(self):
        super().__init__()

        # 윈도우 설정
        self.title("Kiosk Application")
        self.geometry("800x600")

        # 메뉴 항목과 가격 딕셔너리
        self.menu_prices = {
            "커피": {
                "아메리카노": 2000,
                "라떼": 3000,
                "모카": 3000
            },
            "논커피": {
                "유자차": 2500,
                "녹차": 2500,
                "초코라떼": 3000
            }
        }

        # 장바구니 초기화
        self.cart = {}

        # UI 구성
        self.create_widgets()

    def create_widgets(self):
        # 제목 라벨
        self.title_label = ctk.CTkLabel(self, text="Welcome to the Kiosk", font=("Helvetica", 24))
        self.title_label.pack(pady=20)

        # 메뉴 선택 버튼
        self.menu_frame = ctk.CTkFrame(self)
        self.menu_frame.pack(pady=20)

        self.coffee_button = ctk.CTkButton(self.menu_frame, text="커피", command=self.show_coffee_menu)
        self.coffee_button.pack(side="left", padx=10)

        self.noncoffee_button = ctk.CTkButton(self.menu_frame, text="논커피", command=self.show_noncoffee_menu)
        self.noncoffee_button.pack(side="left", padx=10)

        # 메뉴 항목 프레임
        self.menu_items_frame = ctk.CTkFrame(self)
        self.menu_items_frame.pack(pady=20)

        # 장바구니 프레임
        self.cart_frame = ctk.CTkFrame(self)
        self.cart_frame.pack(pady=20)

        self.cart_label = ctk.CTkLabel(self.cart_frame, text="장바구니", font=("Helvetica", 18))
        self.cart_label.pack(pady=10)

        # 장바구니 텍스트 박스 크기 조정
        self.cart_listbox = ctk.CTkTextbox(self.cart_frame, height=80, width=500)
        self.cart_listbox.pack(pady=10)

        self.total_label = ctk.CTkLabel(self.cart_frame, text="총 금액: 0원", font=("Helvetica", 18))
        self.total_label.pack(pady=10)

        # 결과 라벨
        self.result_label = ctk.CTkLabel(self, text="", font=("Helvetica", 18))
        self.result_label.pack(pady=20)

    def show_coffee_menu(self):
        self.clear_menu_items()
        coffee_items = self.menu_prices["커피"]
        for item, price in coffee_items.items():
            button = ctk.CTkButton(self.menu_items_frame, text=f"{item} - {price}원", command=lambda i=item, p=price: self.add_to_cart(i, p))
            button.pack(pady=5)

    def show_noncoffee_menu(self):
        self.clear_menu_items()
        noncoffee_items = self.menu_prices["논커피"]
        for item, price in noncoffee_items.items():
            button = ctk.CTkButton(self.menu_items_frame, text=f"{item} - {price}원", command=lambda i=item, p=price: self.add_to_cart(i, p))
            button.pack(pady=5)

    def clear_menu_items(self):
        for widget in self.menu_items_frame.winfo_children():
            widget.destroy()

    def add_to_cart(self, item, price):
        if item in self.cart:
            self.cart[item]['quantity'] += 1
        else:
            self.cart[item] = {'price': price, 'quantity': 1}
        self.update_cart()

    def remove_from_cart(self, item):
        if item in self.cart:
            del self.cart[item]
            self.update_cart()

    def update_cart(self):
        self.cart_listbox.delete("1.0", ctk.END)
        total_amount = 0
        for item, details in self.cart.items():
            line = f"{item} - {details['quantity']}잔 - {details['price']}원\n"
            self.cart_listbox.insert(ctk.END, line)
            total_amount += details['price'] * details['quantity']
        self.total_label.configure(text=f"총 금액: {total_amount}원")

    def show_selection(self, item):
        for category, items in self.menu_prices.items():
            if item in items:
                price = items[item]
                self.result_label.configure(text=f"선택된 메뉴: {item}\n가격: {price}원")
                break
```

```python
app = KioskApp()
app.mainloop()
```
